```python
import pandas as pd

# 1. In Pandas, we usually create DataFrames from Dictionaries
# The keys become the column names, the lists become the rows
employee_data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40],
    'Salary': [50000, 60000, 75000, 80000]
}

# 2. Convert the dictionary into a Pandas DataFrame
df = pd.DataFrame(employee_data)

# 3. Let's look at our data!
print("Our DataFrame:")
print(df)

# 4. Let's check the shape (Rows, Columns)
print("\nDataFrame Shape:", df.shape)

# 5. Let's look at the column names
print("\nColumn Names:")
print(df.columns)
```
